```python
import logging
from pyspark.sql import functions as F

logging.basicConfig(level=logging.INFO)

try:
    logging.info("Bronze started")

except Exception as e:
    logging.error(str(e))
    raise

```

```python
# reading data from ADLS
df = spark.read.parquet("abfss://fraud@fraudstorageacc1234.dfs.core.windows.net/bronze/transactions/")

logging.info(f"Rows read: {df.count()}")
```

```python
 # adding ingestion time
df = df.withColumn("ingestion_ts", F.current_timestamp())
logging.info("Added ingestion column")
```

```python
# saving Bronze table
df.write \
      .format("delta") \
      .mode("overwrite") \
      .saveAsTable("fraud_catalog.bronze.transactions_raw")

logging.info("Bronze completed")
```
